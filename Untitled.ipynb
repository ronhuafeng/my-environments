```python
import numpy as np
import pandas as pd
from sklearn import svm
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
import imageio
import cv2


image = imageio.imread('test.jpeg')

gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')

print('image shape:', gray.shape)
```

```python
w, h = gray.shape
copy_gray = np.copy(gray)

# 只处理图片的上半部分

bar = 500

data = np.empty(shape=[0, 2])

for i in range(bar):
    for j in range(h):
        if copy_gray[i][j] > 80: # 大于 80 可以认为是白色图像
            copy_gray[i][j] = 255
        else:
            data = np.append(data, [[i, j]], axis=0) # 将上半部分的数据点记录

plt.imshow(copy_gray, cmap='gray')
print('Total points: ', len(data), data.shape)
```

```python
# 使用标志点训练一个 oneclassSVM 分类器

from sklearn.utils import shuffle

shuffle(data)
samples_count = len(data)
points = data[0:samples_count]
print(points.shape)

# Create arbitrary dataset for example
df = pd.DataFrame({'x':     points[:, 1],
                   'y':     points[:, 0],
                   'Label': np.zeros(samples_count, dtype=np.int)}
)

# Fit Support Vector Machine Classifier
X = df[['x', 'y']]
y = df['Label']

clf = svm.OneClassSVM(kernel='rbf', degree=3, gamma=0.1, nu=0.01)
clf.fit(X.values) 
```

```python
# 使用 oneclassSVM 预测每个点是否是在区域内并绘制图形

row, col = gray.shape
copy_gray = np.copy(gray)

for i in range(row):
    for j in range(col):
        if clf.predict([[j, i]])[0] == -1:
            copy_gray[i][j] = 255
        else:
            copy_gray[i][j] = 0

plt.imshow(copy_gray, cmap='gray')
```

```python
import math

row, col = copy_gray.shape

heaps = np.empty(shape=[0, 2])

for j in range(col):
    maxH = None
    for i in range(row):
        if copy_gray[i][j] == 0:
            maxH = i
    if maxH is not None:
        heaps = np.append(heaps, [[j, maxH]], axis=0)        

def distance(p1, p2):
    return math.fabs(p1[0] - p2[0]) + math.fabs(p1[1] - p2[1])

def clean(heaps, threshold=50):
    clean_heaps = np.empty(shape=[0, 2])
    for n, p in enumerate(heaps):
        if n < len(heaps) -1 and n > 0:
            prev = heaps[n-1]
            succ = heaps[n+1]
            if distance(p, prev) > threshold or distance(p, succ) > threshold:
                continue
        clean_heaps = np.append(clean_heaps, [p], axis=0)
    return clean_heaps

clean_heaps = clean(heaps, 50)

plt.imshow(gray, cmap='gray')
plt.plot(clean_heaps[:, 0], clean_heaps[:, 1], marker='o', linestyle='-')
```

```python
clean_heaps = clean(clean_heaps, 50)
clean_heaps = clean(clean_heaps, 50)
clean_heaps = clean(clean_heaps, 50) # can appply clean multiple times



plt.imshow(gray, cmap='gray')
plt.plot(clean_heaps[:, 0], clean_heaps[:, 1], marker='o', linestyle='-')
```
